```python
from bs4 import BeautifulSoup
import requests
import pandas as pd 
import numpy as np
```

```python
def get_title(soup):

    try:

        title = soup.find("span", attrs={'id': 'productTitle'})

        title_string = title.text

        title_value = title_string.strip()

    except AttributeError:
        title_value = ""

    return title_value


def get_price(soup):

    try:

        price = soup.find("span", attrs={'class': 'a-price-whole'})

        price_string = price.text

        price_value = price_string.strip()

    except AttributeError:
        price_value = ""

    return price_value

def get_rating(soup):

    try:

        rating = soup.find("span", attrs={'class': 'a-size-base a-color-base'})

        rating_string = rating.text

        rating_value = rating_string.strip()

    except AttributeError:
        rating_value = ""

    return rating_value

def get_reviews(soup):

    try:

        reviews = soup.find("span", attrs={'id': 'acrCustomerReviewText'})

        reviews_string = reviews.text

        reviews_value = reviews_string.strip()

    except AttributeError:
        reviews_value = ""

    return reviews_value
```

```python
if __name__ == '__main__':

    URL = "https://www.amazon.in/s?k=redmi&ref=nb_sb_noss_1"

    HEADERS = ({'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/112.0.0.0 Safari/537.36', 'Accept-Language': 'en-US, en;q=0.5'})

    webpage = requests.get(URL,headers = HEADERS)


    soup = BeautifulSoup(webpage.content, "html.parser")

    links = soup.find_all("a",attrs = {'class': 'a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal'})

    link_list = []

    for link in links:

        link_list.append(link.get('href'))


    d = {"title":[], "price":[], "rating":[], "reviews":[]}


    for link in link_list:

        new_webpage = requests.get('https://www.amazon.in' + link,headers=HEADERS)


        new_soup = BeautifulSoup(new_webpage.content,"html.parser")

        d['title'].append(get_title(new_soup))
        d['price'].append(get_price(new_soup))
        d['rating'].append(get_rating(new_soup))
        d['reviews'].append(get_reviews(new_soup))

    

```

```python
amazon_df = pd.DataFrame.from_dict(d)
amazon_df['title'].replace('', np.nan, inplace=True)
amazon_df = amazon_df.dropna(subset=['title'])
amazon_df.to_csv("amazon_data.csv", header=True, index=False)
```
